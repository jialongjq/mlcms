```python
system = ["alpha * x1 - x2 - x1 * (x1 ** 2 + x2 ** 2)", "x1 + alpha * x2 - x2 * (x1 ** 2 + x2 ** 2)"]
# showing the phase portrait of steady state behaviour
def draw(alpha):
    phase_portrait_system(system, alpha)

interact(draw, alpha=widgets.FloatSlider(min=-2, max=2, step=0.5, value=-2))
```

```python
ax0 = phase_portrait_system(system, alpha=1)
y0 = np.array([2, 0])
time = np.linspace(0, 10, 100)
yt, time = solve_euler_system(system, alpha=1, y0=y0, time=time)
ax0.plot(yt[:, 0], yt[:, 1], c='red')
ax0.scatter(y0[0], y0[1], c='red')
```
